Chunking with 30 words per chunk and 10 words overlap

```python
import torch
from langchain.text_splitter import CharacterTextSplitter
```

```python
# Step 1: Read the text file
file_path = 'vault.txt'
with open(file_path, 'r', encoding='utf-8') as f:
    text_file = f.read()

```

```python
# Create splitter
splitter = CharacterTextSplitter(
    separator=" ",  # or any custom separator
    chunk_size=1000,  # length of each chunk
    chunk_overlap=500  # overlap between chunks
)

# Split the text
chunks = splitter.split_text(text_file)

print(len(chunks))
```

```python
print(chunks[7])
print(len(chunks))
```

```python
vault_embeddings = []
i=1
for chunk in chunks:
    response = ollama.embeddings(model='nomic-embed-text:latest', prompt=chunk)
    vault_embeddings.append(response["embedding"])
    print(i)
    i+=1
```

```python
data_array=[]
for i in range(len(vault_embeddings)):
    data_array.append({'content':chunks[i],'embedding':vault_embeddings[i]})
```

```python
# To save it to file
file_name='NSU_Code_of_Conduct_2000_1000'
```

```python
import pickle
with open(f'{file_name}.pkl', 'wb') as f:
            pickle.dump(data_array, f)
```

```python
import pickle
vault_embeddings_temp=[]
with open(f'{file_name}.pkl', 'rb') as f:
           vault_embeddings_temp=pickle.load(f)

len(vault_embeddings_temp)
```

```python
db=[]
```

```python


for i in range(len(vault_embeddings_temp)):
    db.append(vault_embeddings_temp[i]['embedding'])
```

```python
# Generate unique IDs for each chunk
ids = [f"chunk_{i+1}" for i in range(len(chunks))]

# Use the chunks themselves as metadata
metadata = [{"text": chunk} for chunk in chunks]
```

```python
# Encode the rewritten input
import torch
```

```python
rewritten_input='Tell me about NSU Financial aid'
top_k=5
input_embedding = ollama.embeddings(model='nomic-embed-text:latest', prompt=rewritten_input)["embedding"]
# Compute cosine similarity between the input and vault embeddings
cos_scores = torch.cosine_similarity(torch.tensor(input_embedding).unsqueeze(0), torch.tensor(db))
# Adjust top_k if it's greater than the number of available scores
top_k = min(top_k, len(cos_scores))
print(top_k)
# Sort the scores and get the top-k indices
top_indices = torch.topk(cos_scores, k=top_k)[1].tolist()
print(top_indices)
# Get the corresponding context from the vault
# relevant_context = [chunk[idx].strip() for idx in top_indices]
relevant_context = [chunks[idx] for idx in top_indices]
for cont in relevant_context:
    print(cont)
    print('===========================')
```
